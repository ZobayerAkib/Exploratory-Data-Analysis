#**Analyzing matches based on the batting-first team's victories provides valuable insights into effective batting strategies and performance trends.**

**Extracting the records where a team won batting first**

```python
Winner_Batting_First=ipl_data[ipl_data['win_by_runs']!=0]
```

```python
Winner_Batting_First.head()
```

```python
#Making a histogram
plt.figure(figsize=(7,7))
plt.hist(Winner_Batting_First['win_by_runs'])
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
Winner_Batting_First['winner'].value_counts()
```

```python
#A bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(12,7))
plt.bar(list(Winner_Batting_First['winner'].value_counts()[0:5].keys()),list(Winner_Batting_First['winner'].value_counts()[0:5]),color=["blue","yellow","orange","Purple","red"])
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(Winner_Batting_First['winner'].value_counts()),labels=list(Winner_Batting_First['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

According to the dataset analysis, the **Mumbai Indians** emerge as the most successful team, having won **57 matches** while **batting first**, closely trailed by the **Chennai Super Kings** with **52** victories.

**The provided plot depicts individual team victories categorized by the run margin.**

```python
# Winner team vs win_by_runs
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(Winner_Batting_First['winner'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(Winner_Batting_First, x='win_by_runs', y='winner', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

#**Analyzing matches based on the batting-Second team's victories provides valuable insights into effective batting strategies and performance trends.**

```python
#looking at the head
batting_second.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
#Making a bar plot for top-3 teams with most wins after batting second
plt.figure(figsize=(12,7))
plt.bar(list(batting_second['winner'].value_counts()[0:5].keys()),list(batting_second['winner'].value_counts()[0:5]),color=["blue","green","orange","brown","red"])
plt.show()
```

**Toss Winner And also Match Winner**

```python
toss_winner_match_winner=ipl_data[ipl_data['toss_winner']==ipl_data['winner']]
toss_winner_match_winner.head()
```

```python
toss_winner_match_winner['winner'].value_counts()
```

```python
plt.figure(figsize=(12,7))
plt.bar(list(toss_winner_match_winner['winner'].value_counts()[0:5].keys()),list(toss_winner_match_winner['winner'].value_counts()[0:5]),color=["blue","green","orange","brown","red"])
plt.title("Team vs (Match Winner and Toss Winner both)")
plt.xlabel("Winner Team")
plt.ylabel("Number of Matches")
plt.show()
```

The data indicates that **Chennai Super Kings (CSK)** won **57 matches**, while **Mumbai Indians (MI)** won **56 matches**. Additionally, it specifies that both teams were also the **toss winners** in these matches.

**Toss Winner and Match Winner And Batting First**

```python
TMB_Winner=toss_winner_match_winner[toss_winner_match_winner['win_by_runs']!=0]
TMB_Winner.head()
```

```python
TMB_Winner['winner'].value_counts()
```

```python
plt.figure(figsize=(12,7))
x=list(TMB_Winner['winner'].value_counts()[0:5].keys())
y=list(TMB_Winner['winner'].value_counts()[0:5])
plt.bar(x,y,color=["blue","green","orange","brown","red"])
plt.title("Team vs (Match Winner, Toss Winner and Batting First)")
plt.xlabel("Winner Team")
plt.ylabel("Number of Matches")
plt.show()
```



**Chennai Super Kings** won **30 matches** and **Mumbai Indians** won **24 matches** where the toss winner also chose to **bat first** and emerged as the **match winner**. **This indicates a strategic advantage in winning the toss and opting to bat first.**
